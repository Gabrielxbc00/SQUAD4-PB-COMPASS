```python
from pyspark.sql import SparkSession
```

```python
contagem = SparkSession.builder.appName("Contagem de Palavras").getOrCreate()
```

```python
arquivo = "README.md"
```

```python
rdd = contagem.sparkContext.textFile(arquivo)
```

```python
contagem_palavras = rdd.flatMap(lambda line: line.split(" ")).countByValue()
```

```python
for word, count in contagem_palavras.items():
    print(f"{word}: {count}")
```
